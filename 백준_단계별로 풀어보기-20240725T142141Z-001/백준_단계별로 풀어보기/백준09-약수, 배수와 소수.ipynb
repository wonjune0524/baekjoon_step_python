```python
# 01

while True:
  A, B = map(int, input().split())

  if A == 0 and B == 0:
    break
  elif B % A == 0:
    print("factor")
  elif A % B == 0:
    print("multiple")
  else:
    print("neither")
```

```python
# 02

N, K = map(int, input().split())
N_measure = []

for i in range(N):
  if N % (i + 1) == 0:
    N_measure.append(i + 1)

if K > len(N_measure):
  print(0)
else:
  print(N_measure[K - 1])
```

```python
# 03

while True:
  n = int(input())
  n_measure = []
  n_sum = 0

  if n == -1:
    break

  for i in range(n - 1):
    if n % (i + 1) == 0:
      n_measure.append(i + 1)

  for i in range(len(n_measure)):
    n_sum += n_measure[i]

  if n == n_sum:
    print(f"{n} = ", end = "")
    print(" + ".join(map(str, n_measure)))
  else:
    print(f"{n} is NOT perfect.")w
```

```python
# 04

N = int(input())
num_list = list(map(int, input().split()))
result = 0

for i in num_list:
  count = 0

  for j in range(i):
    if i % (j + 1) == 0:
      count += 1

  if count == 2:
    result += 1

print(result)
```

```python
# 05

N = int(input())
M = int(input())
num_sum = 0
num_min = 0

for i in range(N, M + 1):
  count = 0

  for j in range(i):
    if i % (j + 1) == 0:
      count += 1

  if count == 2:
    if num_sum == 0:
      num_min = i
    num_sum += i

if num_min == 0:
  print(-1)
else:
  print(num_sum)
  print(num_min)
```

```python
# 05 - 시간 초과로 인한 다른 방법 풀이

N = int(input())
M = int(input())
num_list = []

for i in range(N, M + 1):
  count = 0
  if i > 1:
    for j in range(2, i):
      if i % j == 0:
        count += 1
        break

    if count == 0:
      num_list.append(i)

if len(num_list) < 1:
  print(-1)
else:
  print(sum(num_list))
  print(min(num_list))
```

```python
# 06

N = int(input())
num_list = []

while N != 1:
  for i in range(1, N):
    if N % (i + 1) == 0:
      num_list.append(i + 1)
      N //= (i + 1)
      break

for i in num_list:
  print(i)
```
