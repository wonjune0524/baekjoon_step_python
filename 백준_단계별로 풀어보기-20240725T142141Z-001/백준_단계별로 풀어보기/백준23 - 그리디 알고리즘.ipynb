```python
# 01

N, K = map(int, input().split())
coin = [int(input()) for _ in range(N)]
coin_count = 0

for i in range(1, (N + 1)):
  coin_count += K // coin[-i]
  K %= coin[-i]

print(coin_count)
```

```python
# 02

N = int(input())
time = [list(map(int, input().split())) for _ in range(N)]

time.sort(key = lambda x : (x[1], x[0]))

count_time = 1
end_time = time[0][1]

for i in range(1, N):
  if end_time <= time[i][0]:
    end_time = time[i][1]
    count_time += 1

print(count_time)
```

```python
# 03

N = int(input())
wd_time = list(map(int, input().split()))
prefix_sum = [0] * (N + 1)

wd_time.sort()

for i in range(1, (N + 1)):
  prefix_sum[i] = prefix_sum[i - 1] + wd_time[i - 1]
print(sum(prefix_sum))
```

```python
# 04

# 식 입력
expression = input().strip()

# '-'를 기준으로 식을 나눔
parts = expression.split('-')

# 첫 번째 부분은 더하고 나머지는 모두 빼기
result = sum(map(int, parts[0].split('+')))

for part in parts[1:]:
  result -= sum(map(int, part.split('+')))

print(result)
```

```python
# 04 - 다시 풀어보기
s = input().strip()

parts = s.split('-')

result = sum(map(int, parts[0].split('+')))

for part in parts[1:]:
  result -= sum(map(int, part.split('+')))
print(result)
```

```python
# 05

N = int(input())
distance = list(map(int, input().split()))
oil_cost = list(map(int, input().split()))

total_cost = distance[0] * oil_cost[0]
min_city = oil_cost[0]
for i in range(1, (N - 1)):
  if min_city <= oil_cost[i]:
    total_cost += distance[i] * min_city
  else:
    min_city = oil_cost[i]
    total_cost += distance[i] * min_city

print(total_cost)
```
