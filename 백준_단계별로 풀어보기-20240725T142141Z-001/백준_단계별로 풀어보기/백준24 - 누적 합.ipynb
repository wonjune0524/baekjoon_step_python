```python
# 01

N, M = map(int, input().split())
num_list = list(map(int, input().split()))
dp = [0] * M

for num in range(M):
  i, j = map(int, input().split())

  for index in range(i - 1, j):
    dp[num] += num_list[index]

for i in dp:
  print(i)
```

```python
# 01 - 시간 초과로 인한 다른 풀이 / 누적 합 이용

N, M = map(int, input().split())
num_list = list(map(int, input().split()))

prefix_sum = [0] * (N + 1)
for i in range(1, N + 1):
  prefix_sum[i] = prefix_sum[i - 1] + num_list[i - 1]

results = []
for _ in range(M):
  i, j = map(int, input().split())
  results.append(prefix_sum[j] - prefix_sum[i - 1])

print('\n'.join(map(str, results)))
```

```python
# 01 - 다시 풀어보기

N, M = map(int, input().split())
num_list = list(map(int, input().split()))

prefix_sum = [0] * (N + 1)
for i in range(1, N + 1):
  prefix_sum[i] = prefix_sum[i - 1] + num_list[i - 1]

results = []
for _ in range(M):
  i, j = map(int, input().split())
  results.append(prefix_sum[j] - prefix_sum[i - 1])

print('\n'.join(map(str, results)))
```

```python
# 02

N, K = map(int, input().split())
temp = list(map(int, input().split()))

prefix_sum = [0] * (N + 1)
for i in range(1, N + 1):
  prefix_sum[i] = prefix_sum[i - 1] + temp[i - 1]

results = []
for i in range(N - K + 1):
  results.append(prefix_sum[i + K] - prefix_sum[i])

print(max(results))
```

```python
# 03

S = input()
q = int(input())

for i in range(q):
  alp, start, end = input().split()
  start = int(start)
  end = int(end)

  prefix_sum = [0] * (len(S) + 1)
  for i in range(1, len(S) + 1):
    if alp == S[i - 1]:
      prefix_sum[i] = prefix_sum[i - 1] + 1
    else:
      prefix_sum[i] = prefix_sum[i - 1]

  print(prefix_sum[end + 1] - prefix_sum[start])
```

```python
# 03 - 더 좋은 방법

s = input().rstrip()
q = int(input().rstrip())

count = {0 : [0] * 26}

for i, ch in enumerate(s):
  count[i + 1] = count[len(count) - 1][:]
  count[i + 1][ord(ch) - 97] += 1

for _ in range(q):
  alpha, l, r = input().rstrip().split()
  answer = count[int(r) + 1][ord(alpha) - 97] - count[int(l)][ord(alpha) - 97]
  print(answer)
print(count)
```

```python
# 04
N, M = map(int, input().split())
num = list(map(int, input().split()))
count = 0

prefix_sum = [0] * (N + 1)
for i in range(1, N + 1):
  prefix_sum[i] += prefix_sum[i - 1] + num[i - 1]

for i in range(N):
  for j in range(i + 1, N + 1):
    if (prefix_sum[j] - prefix_sum[j - (i + 1)]) % M == 0:
      count += 1

print(count)
```

```python
# 04 - 시간 초과로 인한 다른 풀이

N, M = map(int, input().split())
a = list(map(int, input().split()))

r = [0] * M   # M으로 나눈 나머지 담는 곳
prefixSum = 0
for i in range(N):
  prefixSum += a[i]
  r[prefixSum % M] += 1

ans = r[0]   # 나머지가 0이 되는 경우의 수
for i in range(M):
  # nC2 = n(n - 1) / 2
  ans += r[i] * (r[i] - 1) // 2
print(ans)
```

```python
# 04 - 다시 풀어보기

N, M = map(int, input().split())
num = list(map(int, input().split()))

r = [0] * M
prefixSum = 0
for i in range(N):
  prefixSum += num[i]
  r[prefixSum % M] += 1

ans = r[0]
for i in range(M):
  ans += r[i] * (r[i] - 1) // 2

print(ans)
```

```python
# 05 - 시간 초과로 인한 다른 풀이

N, M = map(int, input().split())

arr = []
for i in range(N):
  arr.append(list(map(int, input().split())))

prefix_sum = [[0] * (N + 1) for _ in range(N + 1)]

for i in range(N):
  for j in range(N):
    prefix_sum[i + 1][j + 1] = prefix_sum[i][j + 1] + prefix_sum[i + 1][j] - prefix_sum[i][j] + arr[i][j]

for i in range(M):
  x1, y1, x2, y2 = map(int, input().split())

  sum = prefix_sum[x2][y2] - prefix_sum[x1 - 1][y2] - prefix_sum[x2][y1 - 1] + prefix_sum[x1 - 1][y1 - 1]
  print(sum)
```

```python
# 05 - 다시 풀어보기

N, M = map(int, input().split())
arr = []

for i in range(N):
  arr.append(list(map(int, input().split())))

prefix_sum = [[0] * (N + 1) for _ in range(N + 1)]
for i in range(1, N + 1):
  for j in range(1, N + 1):
    prefix_sum[i][j] = prefix_sum[i - 1][j] + prefix_sum[i][j - 1] - prefix_sum[i - 1][j - 1] + arr[i - 1][j - 1]

for i in range(M):
  x1, y1, x2, y2 = map(int, input().split())
  result = prefix_sum[x2][y2] - prefix_sum[x1 - 1][y2] - prefix_sum[x2][y1 - 1] + prefix_sum[x1 - 1][y1 - 1]
  print(result)
```

```python
# 06

N, M, K = map(int, input().split())

board_list = []
count_list = []
for i in range(N):
  board_list.append(input())

for a in range(N - (K - 1)):
  for b in range(M - (K - 1)):
    B_start_count = 0
    W_start_count = 0
    for i in range(a, a + K):
      for j in range(b, b + K):
        if (i + j) % 2 == 0:
          if board_list[i][j] != 'B':
            B_start_count += 1
          elif board_list[i][j] != 'W':
            W_start_count += 1
        else:
          if board_list[i][j] != 'B':
            W_start_count += 1
          elif board_list[i][j] != 'W':
            B_start_count += 1

    count_list.append(B_start_count)
    count_list.append(W_start_count)

print(min(count_list))
```

```python
# 06 - 시간 초과로 인한 다른 풀이

N, M, K = map(int, input().split())
board_list = [input().rstrip() for _ in range(N)]

prefix_W_sum = [[0] * (M + 1) for _ in range(N + 1)]
prefix_B_sum = [[0] * (M + 1) for _ in range(N + 1)]

for i in range(1, N + 1):
  for j in range(1, M + 1):
    if (i + j) % 2 == 0:
      count_W = 0 if board_list[i - 1][j - 1] == 'W' else 1
      count_B = 0 if board_list[i - 1][j - 1] == 'B' else 1
    else:
      count_W = 0 if board_list[i - 1][j - 1] == 'B' else 1
      count_B = 0 if board_list[i - 1][j - 1] == 'W' else 1

    prefix_W_sum[i][j] = prefix_W_sum[i - 1][j] + prefix_W_sum[i][j - 1] - prefix_W_sum[i - 1][j - 1] + count_W
    prefix_B_sum[i][j] = prefix_B_sum[i - 1][j] + prefix_B_sum[i][j - 1] - prefix_B_sum[i - 1][j - 1] + count_B

min_count = 2000000
for a in range(K, N + 1):
  for b in range(K, M + 1):
    count_W = prefix_W_sum[a][b] - prefix_W_sum[a - K][b] - prefix_W_sum[a][b - K] + prefix_W_sum[a - K][b - K]
    count_B = prefix_B_sum[a][b] - prefix_B_sum[a - K][b] - prefix_B_sum[a][b - K] + prefix_B_sum[a - K][b - K]
    min_count = min(min_count, count_W, count_B)

print(min_count)
```

```python
# 06 - 다시 풀어보기

N, M, K = map(int, input().split())
board_list = [input().strip() for _ in range(N)]

# W로 시작
prefix_W_sum = [[0] * (M + 1) for _ in range(N + 1)]
# B로 시작
prefix_B_sum = [[0] * (M + 1) for _ in range(N + 1)]

# 누적합 배열 채우기
for i in range(1, (N + 1)):
  for j in range(1, (M + 1)):
    if (i + j) % 2 == 0:
      count_W = 0 if board_list[i - 1][j - 1] == 'W' else 1
      count_B = 0 if board_list[i - 1][j - 1] == 'B' else 1
    else:
      count_W = 0 if board_list[i - 1][j - 1] == 'B' else 1
      count_B = 0 if board_list[i - 1][j - 1] == 'W' else 1
    prefix_W_sum[i][j] = prefix_W_sum[i - 1][j] + prefix_W_sum[i][j - 1] - prefix_W_sum[i - 1][j - 1] + count_W
    prefix_B_sum[i][j] = prefix_B_sum[i - 1][j] + prefix_B_sum[i][j - 1] - prefix_B_sum[i - 1][j - 1] + count_B

min_count = 2000000
for a in range(K, (N + 1)):
  for b in range(K, (M + 1)):
    count_W = prefix_W_sum[a][b] - prefix_W_sum[a][b - K] - prefix_W_sum[a - K][b] + prefix_W_sum[a - K][b - K]
    count_B = prefix_B_sum[a][b] - prefix_B_sum[a][b - K] - prefix_B_sum[a - K][b] + prefix_B_sum[a - K][b - K]
    min_count = min(min_count, count_W, count_B)

print(min_count)
```
