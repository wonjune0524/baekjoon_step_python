```python
# 01

word, num = input().split()
sum_num = 0

for i in range(len(word)):
  if ord(word[i]) >= 65:   # 알파벳 대문자일 때
    sum_num += (int(num) ** (len(word) - 1 - i)) * (ord(word[i]) - 55)
  else:   # 숫자일 때
    sum_num += (int(num) ** (len(word) - 1 - i)) * (ord(word[i]) - 48)

print(sum_num)
```

```python
# 01 - 다른 풀이방법

num_list = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

n, b = input().split()
sum = 0
for i, num in enumerate(n[::-1]):
    sum += (int(b) ** i) * num_list.index(num)
print(sum)
```

```python
# 02 - 이는 매우 복잡하므로 실패.

alp_list = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

N, B = map(int, input().split())
num_list = []
count = 0
num_list.append(alp_list[N % B])
N -= N % B
fst_B = B

while True:
  count += 1
  B = fst_B ** count

  if N // B >= fst_B:
    num_list.append(alp_list[N // B % fst_B])
    N -= (N // B % fst_B) * B
  else:
    num_list.append(alp_list[N // B])
    break


for i in range(len(num_list), 0, -1):
  print(num_list[i - 1], end ="")   # 리스트를 벗어나 역순 추출해야함
```

```python
# 02 - 훨씬 간단하게 풀어보기

alp_list = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
answer = ""
N, B = map(int, input().split())

while N != 0:
  answer = answer + alp_list[N % B]
  N //= B

for i in range(len(answer) -1, -1, -1):
  print(answer[i], end ="")
```

```python
# 03

change_type = [25, 10, 5, 1]
T = int(input())

for i in range(T):
  change_list = []
  pay = int(input())

  for j in range(len(change_type)):
    change_list.append(pay // change_type[j])
    pay %= change_type[j]

  for k in range(len(change_list)):
    print(change_list[k], end = " ")

  print()
```

```python
# 05

N = int(input())
count = 0

while N > 0:
  if count == 0:
    N -= 1
  else:
    N -= 6 * count
  count += 1

print(count)
```

```python
# 06

X = int(input())
index = 0

while X > 0:
  index += 1
  X_last = X
  X -= index

if index == 1 or index % 2 == 0:
  print(f"{X_last}/{index + 1 - X_last}")
else:
  print(f"{index + 1 - X_last}/{X_last}")
```

```python
# 07

A, B, V = map(int, input().split())
climb_day = 0
climb_height = 0

while climb_height < V:
  climb_day += 1
  climb_height += A

  if climb_height >= V:
    break
  else:
    climb_height -= B

print(climb_day)
```

```python
# 07 - 시간 초과로 인한 다른 풀이 방법

A, B, V = map(int, input().split())

if (V - A) % (A - B) != 0:
  print((V - A) // (A - B) + 2)
else:
  print((V - A) // (A -B) + 1)
```
