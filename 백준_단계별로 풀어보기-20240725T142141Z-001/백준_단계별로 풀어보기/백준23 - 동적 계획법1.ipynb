```python
# 01

def fib(num):
  if num <= 1:
    return num
  return fib(num - 1) + fib(num - 2)

def fibonacci(num):
  f = [0] * (num + 1)
  f[1] = f[2] = 1

  for i in range(3, num + 1):
    f[i] = f[i - 1] + f[i - 2]
  return f[num]

n = int(input())
fib_result = fib(n)
print(fib_result, n - 2)
```

```python
# 02

w = [[[0 for _ in range(21)] for _ in range(21)] for _ in range(21)]
# w_list = [[[1] * 21] * 21] + [[[1] * 21] + [[1] + [0] * 20] * 20] * 20
# 이는 얕은 복사이므로 좋은 방법이 아님

def abc(a, b, c):
  if a <= 0 or b <= 0 or c <= 0:
    return 1

  if a > 20 or b > 20 or c > 20:
    return abc(20, 20, 20)

  if w[a][b][c]:
    return w[a][b][c]

  if a < b and b < c:
    w[a][b][c] =  abc(a, b, c - 1) + abc(a, b - 1, c - 1) - abc(a, b - 1, c)
    return w[a][b][c]

  else:
    w[a][b][c] = abc(a - 1, b, c) + abc(a - 1, b - 1, c) + abc(a - 1, b, c - 1) - abc(a - 1, b - 1, c - 1)
    return w[a][b][c]

while True:
  a, b, c = map(int, input().split())
  if a == b == c == -1:
    break
  print(f"w({a}, {b}, {c}) = {abc(a, b, c)}")
```

```python
# 05 - 동적 계획법 고려하지 않고 품 / 이는 시간 초과가 발생함

n = int(input())
num_list = list(map(int, input().split()))
max_list = []

for i in range(n):
  num_sum = num_list[i]
  for j in range(i, n):
    if j != i:
      num_sum += num_list[j]

    if not max_list:
      max_list.append(num_sum)

    if num_sum > max_list[0]:
      max_list[0] = num_sum

print(max_list[0])
```

```python
# 05 - 동적 계획법 활용(카데인 알고리즘 사용)

n = int(input())
num_list = list(map(int, input().split()))

# 첫 번째 값 초기화
current_max = num_list[0]
global_max = num_list[0]

# 두 번째 값부터 끝까지 처리
for i in range(1, n):
  current_max = max(num_list[i], current_max + num_list[i])   # 뒤의 값보다 앞의 num_list[i]값이 더 클 경우, 기존에 더한 모든 값을 초기화시키고 i index부터 다시 시작한다.
  global_max = max(global_max, current_max)     # global_max는 current_max에 num_list[i]를 더하기 전의 값이다. 이 값도 고려해줘야 하기 때문에 global_max에 값을 저장해둔 후 비교하는 것임.

print(global_max)
```

```python
# 06 - 동적 계획법(Dynamic Programming) 사용

N = int(input())
price_list = [list(map(int, input().split())) for _ in range(N)]

dp = [[0] * 3 for _ in range(N)]

dp[0][0] = price_list[0][0]   # 첫 번째 집 빨강 시작
dp[0][1] = price_list[0][1]   # 첫 번째 집 초록 시작
dp[0][2] = price_list[0][2]   # 첫 번째 집 파랑 시작

for i in range(N):
  dp[i][0] = min(dp[i - 1][1], dp[i - 1][2]) + price_list[i][0]   # 빨강으로 칠했을 때
  dp[i][1] = min(dp[i - 1][0], dp[i - 1][2]) + price_list[i][1]   # 초록으로 칠했을 때
  dp[i][2] = min(dp[i - 1][0], dp[i - 1][1]) + price_list[i][2]   # 파랑으로 칠했을 때

print(min(dp[N - 1][0], dp[N - 1][1], dp[N - 1][2]))
```

```python
# 07

n = int(input())
tri_list = [list(map(int, input().split())) for _ in range(n)]
sum_list = [[0] * i for i in range(1, n + 1)]

sum_list[0][0] = tri_list[0][0]

for i in range(1, n):
  for j in range(len(sum_list[i])):
    if j == 0:
      sum_list[i][j] = sum_list[i - 1][j] + tri_list[i][j]
    elif j == len(sum_list[i]) - 1:
      sum_list[i][j] = sum_list[i - 1][j - 1] + tri_list[i][j]
    else:
      sum_list[i][j] = max(sum_list[i - 1][j - 1], sum_list[i - 1][j]) + tri_list[i][j]

print(max(sum_list[n - 1]))
```

```python
# 08 - 틀린 부분 수정(또 틀림)

N = int(input())
stairs = [int(input()) for _ in range(N)]

if N == 1:
  print(stairs[0])
  exit()
elif N == 2:
  print(stairs[0] + stairs[1])
  exit()

score = [0] * N   # 각 계단에 도달하는 점수의 최댓값을 저장하는 리스트

score[0] = stairs[0]
score[1] = stairs[0] + stairs[1]
score[2] = max(stairs[0] + stairs[2], stairs[1] + stairs[2])

for i in range(3, N):
  if i % 2 == 0:
    score[i] = max(score[i - 2] + stairs[i], score[i - 1])
  else:
    score[i] = max(score[i - 2] + stairs[i], score[i - 3] + stairs[i - 1] + stairs[i])
    # 앞의 값은 12칸을 의미, 뒤의 값은 21칸을 의미함.

print(score[N - 1])
```

```python
# 08 - 다른 풀이

n = int(input())  # 계단 개수 입력
stairs = [int(input()) for _ in range(n)]  # 각 계단의 점수 입력

# 계단이 한 개거나 두 개만 있을 때의 예외 처리
if n == 1:
    print(stairs[0])
elif n == 2:
    print(stairs[0] + stairs[1])
else:
    # dp 테이블 초기화
    dp = [0] * n

    # 첫 번째 계단까지의 최댓값은 첫 번째 계단의 점수
    dp[0] = stairs[0]
    # 두 번째 계단까지의 최댓값은 첫 번째 + 두 번째 계단의 점수
    dp[1] = stairs[0] + stairs[1]
    # 세 번째 계단까지의 최댓값은 두 번째 또는 첫 번째에서 세 번째로 갈 수 있음
    dp[2] = max(stairs[1] + stairs[2], stairs[0] + stairs[2])

    # 네 번째 계단부터는 규칙에 따라 dp를 채움
    for i in range(3, n):
        dp[i] = max(dp[i-2] + stairs[i], dp[i-3] + stairs[i-1] + stairs[i])

    # 마지막 계단까지의 최댓값 출력
    print(dp[n-1])
```

```python
# 09 - 이는 Greedy(탐욕적) 방식이라 비효율적일 수 있음. 그리고 백준에서 틀림

X = int(input())
count_X = 0

while X != 1:
  if (X % 3 != 0 and X % 2 != 0) or (X - 1) % 3 == 0:
    X -= 1
  elif X % 3 == 0:
    X //= 3
  elif X % 3 != 0 and X % 2 == 0:
    X //= 2
  count_X += 1

print(count_X)
```

```python
# 10

N = int(input())
MOD = 1000000000      # modulus : 나머지

dp = [[0] * 10 for i in range(N + 1)]   # 맨 마지막 자릿수 0~9 에 해당하는 숫자들 가짓수를 나타내기 위한 리스트

for i in range(1, 10):
  dp[1][i] = 1

for length in range(2, N + 1):
  for digit in range(10):
    if digit == 0:
      dp[length][digit] = dp[length - 1][1] % MOD
    elif digit == 9:
      dp[length][digit] = dp[length - 1][8] % MOD
    else:
      dp[length][digit] = (dp[length - 1][digit - 1] + dp[length - 1][digit + 1]) % MOD

result = sum(dp[N]) % MOD
print(result)
```

```python
# 13

N = int(input())
A = list(map(int, input().split()))
dp1 = [1] * N
dp2 = [1] * N

for i in range(1, N):
  for j in range(i):
    if A[j] < A[i]:
      dp1[i] = max(dp1[i], dp1[j] + 1)

for i in range(N - 2, -1, -1):
  for j in range(N - 1, i, -1):
    if A[i] > A[j]:
      dp2[i] = max(dp2[i], dp2[j] + 1)

result = [(dp1[i] + dp2[i] - 1) for i in range(N)]
print(max(result))
```

```python
# 14 - LIS(Longest Increasing Subsequence) 사용 / 최장 증가 부분 수열

N = int(input())
elec = [list(map(int, input().split())) for _ in range(N)]
elec.sort()       # 순서대로 처리하기 위해 오름차순 정렬
dp = [1] * N      # 마지막 번호까지 배열을 줄 필요없음. 전깃줄 개수만큼만 고려하자.

for i in range(1, N):
  for j in range(i):
    if elec[i][1] > elec[j][1]:
      dp[i] = max(dp[i], dp[j] + 1)

print(N - max(dp))
```

- [1] 1번은 0자리가 자기보다 크고, 1자리가 자기보다 작을 때 교차함.
- [2] 마지막번은 0자리가 자기보다 작고, 1자리가 자기보다 클 때 교차함.
- [3] 나머지는 (1) 0자리가 자기보다 작을 때, 1자리가 자기보다 커야 함. (2) 0자리가 자기보다 클 때, 1자리가 자기보다 작아야 함.

- 전깃줄이 교차하지 않도록 만들기 위해서는 A 전봇대의 순서에 맞춰 B 전봇대의 연결도 순서대로 증가해야 한다. 즉, A 전봇대에서의 번호가 증가할 때, B 전봇대의 번호도 증가해야 교차하지 않는 것이다.

- 따라서, A 전봇대 기준으로 B 전봇대의 번호들을 살펴보고, 그 안에서 최장 증가하는 부분을 찾아야 한다. 이 "증가하는 부분"이 바로 전깃줄이 교차하지 않는 전깃줄의 개수를 의미한다.

```python
# 15 - LCS(Longest Common Subsequence) 사용 문제 / 최장 공통 부분 수열

lcs1 = input()
lcs2 = input()
dp = [[0] * (len(lcs2) + 1) for _ in range(len(lcs1) + 1)]    # 문자열이 비어있을 때 처리할 수 있어야 하므로 길이에 +1 을 해줌.

for i in range(1, len(lcs1) + 1):
  for j in range(1, len(lcs2) + 1):
    if lcs1[i - 1] == lcs2[j - 1]:    # 문자가 같을 때
      dp[i][j] = dp[i - 1][j - 1] + 1
    else:   # 문자가 다를 때
      dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

print(dp[-1][-1])
```

- 둘 중 하나의 문자열을 고정하자(긴지 짧은지는 상관없기에 그냥 첫 번째 문자열을 고정하기로 했음). 각 문자에 대해서 부분 수열의 길이를 계산하여 dp 리스트에 채워넣자.
- 첫 번째 문자열의 문자(a) 하나를 고정한다. 두 번째 문자열을 반복문을 돌리면서 문자(a)와 같은 문자가 있으면 dp 리스트에 +1(이는 부분 수열의 길이가 1 증가함을 의미함)을 해주면서 문자(a)의 다음 문자로 넘어가준다.

```python
# 16

N, K = map(int, input().split())
backpack = [list(map(int, input().split())) for _ in range(N)]
dp = [0] * (K + 1)

for weight, value in backpack:
  for j in range(K, weight - 1, -1):
    dp[j] = max(dp[j], dp[j - weight] + value)

print(dp)
```
