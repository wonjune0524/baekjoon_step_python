```python
# 01

N, M = map(int, input().split())

def prcs():   # 행렬 생성 함수
  arr_list = []   # 또는 list()
  for i in range(N):
    a = input().split()
    if len(a) == M:
      arr_list.append(a)
  return arr_list

A = prcs()
B = prcs()

for i in range(N):
  for j in range(M):
    print(int(A[i][j]) + int(B[i][j]), end=" ")
  print("")
```

```python
# 02

num_list = []

for i in range(9):    # 모든 값을 이중 리스트에 집어넣기
  a = list(map(int, input().split()))
  if len(a) == 9:
    num_list.append(a)

max_value = num_list[0][0]
index_i = 0
index_j = 0

for i in range(9):
  for j in range(9):
    if num_list[i][j] > max_value:
      max_value = num_list[i][j]
      index_i = i
      index_j = j

print(max_value)
print(f"{index_i + 1} {index_j + 1}")
```

```python
# 03

str_list = []

for i in range(5):    # 이중 리스트에 문자열 배열
  a = input()
  str_list.append(a)

max_len = len(str_list[0])

for i in range(5):
  if len(str_list[i]) > max_len:
    max_len = len(str_list[i])

for i in range(max_len):
  for j in range(5):
    if len(str_list[j]) > i:
      print(str_list[j][i], end = "")
    else:
      pass    # 값을 출력하지 않고 넘어가게 함
```

```python
# 03 - max_len 다른 방법 / 왜 오류뜨는지 모르겠는데 코드는 맞음

str_list = []

for i in range(5):    # 이중 리스트에 문자열 배열
  a = input()
  str_list.append(a)

max_len = max(len(s) for s in str_list)

for i in range(max_len):
  for j in range(5):
    if len(str_list[j]) > i:
      print(str_list[j][i], end = "")
    else:
      pass    # 값을 출력하지 않고 넘어가게 함
```

```python
# 04

N = int(input())
count_num = 0
white_map = [[0 for i in range(100)] for j in range(100)]   # (99,99) 까지 찍음. 점을 넓이가 1인 정사각형의 왼쪽 아래 모서리에 찍는다고 생각.

for i in range(N):    # 점 찍기 반복문
  x, y = map(int, input().split())

  for j in range(x, x + 10):
    for k in range(y, y + 10):
      if white_map[j][k] == 0:
        white_map[j][k] = 1
      else:
        pass

for arr_num in white_map:
  count_num += arr_num.count(1)

print(count_num)
```
