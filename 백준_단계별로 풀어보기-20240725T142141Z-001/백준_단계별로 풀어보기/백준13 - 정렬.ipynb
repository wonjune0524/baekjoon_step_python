```python
# 05 - 계수 정렬(counting sort)

import sys
input = sys.stdin.readline

# 계수정렬 활용
N = int(input())
arr = [0] * (10000 + 1) # 입력값이 10000개까지 주어지니 10000 + 1개의 리스트 선언

# 각 입력값에 해당하는 인덱스의 값 증가
for _ in range(N):
    arr[int(input())] += 1

# arr에 기록된 정보 확인
for i in range(len(arr)):
    if arr[i] != 0: # 0이 아닌 데이터, 즉 입력받은 데이터들을 출력
        for _ in range(arr[i]):
            print(i)
```

```python
# 07

N = int(input())
x_list = []
y_list  = []
count = 0

for i in range(N):
  x, y = map(int, input().split())
  x_list.append(x)
  y_list.append(y)
  count += 1

x_list.sort()
y_list.sort()

for i in range(N - 1):
  for j in range(i + 1, N):
    if x_list[i] == x_list[j]:
      if y_list[i] > y_list[j]:
        y_list[i], y_list[j] = y_list[j], y_list[i]

for i in range(N):
  print(f"{x_list[i]} {y_list[i]}")
```

```python
# 07 - 시간 초과로 인한 다른 풀이

N = int(input())
arr_list = []

for i in range(N):
  x, y = map(int, input().split())
  arr_list.append((x, y))

arr_list.sort()

for i in range(N):
  print(arr_list[i][0], arr_list[i][1])
```

```python
# 07 - 다른 풀이 방법

N = int(input())
arr_list = []

for i in range(N):
  x, y = map(int, input().split())
  arr_list.append((x, y))

arr_list.sort(key = lambda x : (x[0], x[1]))

for i in range(N):
  print(arr_list[i][0], arr_list[i][1])
```

```python
# 09 - 시간초과 가능성이 있기에 아래 다른 방법 먼저

N = int(input())
word_list = []

for i in range(N):
  word = input()
  if word not in word_list:
    word_list.append(word)

word_list.sort()

for i in range(len(word_list) - 1):
  for j in range(i + 1, len(word_list)):
    if len(word_list[j]) < len(word_list[i]):
      word_list[i], word_list[j] = word_list[j], word_list[i]

for i in word_list:
  print(i)
```

```python
# 09

N = int(input())
word_list = []

for i in range(N):
  word = input()
  if word not in word_list:
    word_list.append(word)

word_list.sort(key = lambda word : (len(word), word))

for i in word_list:
  print(i)
```

```python
# 10

N = int(input())
member_list = []

for i in range(N):
  age, name = input().split()
  member_list.append((int(age), name))   # 두 요소를 괄호로 반드시 묶어야함.

member_list.sort(key = lambda mem : mem[0])

for i in member_list:
  print(i[0], i[1])
```

```python
# 11 - 시간 초과로 인한 다른 방법

N = int(input())
num_list = list(map(int, input().split()))
num_set = sorted(list(set(num_list)))

dic = {}
for i in range(len(num_set)):
  dic[num_set[i]] = i

for i in num_list:
  print(dic[i], end=" ")
```
