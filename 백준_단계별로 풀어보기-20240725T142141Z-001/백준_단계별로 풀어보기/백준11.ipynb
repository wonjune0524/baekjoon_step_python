```python
# 04 - 런타임 에러로 인해 다시 풀기

n = int(input())
sum = 0

for i in range(n - 1):
  for j in range(i + 1, n):
    sum += 1

print(sum)
print(2)
```

```python
# 04 - 시간 초과로 인해 다시 풀기

n = int(input())

print(int(n * (n - 1) - n * (n - 1) / 2))
print(2)
```

```python
# 07

a1, a0 = map(int, input().split())
c = int(input())
n0 = int(input())

if a1 <= c and a1 * n0 + a0 <= c * n0:
  print(1)
else:
  print(0)
```
