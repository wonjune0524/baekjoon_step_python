```python
# 04
def cut(a, b, c):
  # b가 1이 아닐 때, 재귀
  if b != 1:
    if b % 2 == 1:
      left_cut = cut(a, b // 2, c)
      right_cut = cut(a, b // 2 + 1, c)
      return left_cut * right_cut % c
    else:
      left_cut = cut(a, b // 2, c)
      return left_cut ** 2 % c

  # b가 1일 때
  return a % c

A, B, C = map(int, input().split())
print(cut(A, B, C))
```

# 나머지 연산의 분배법칙
- (A + B) % p = ((A % p) + (B % p)) % p
- (A x B) % p = ((A % p) x (B % p)) % p
- (A - B) % p = ((A % p) - (B % p) + p) % p

```python
# 07
# 행렬 곱 함수
def cal(arr1, arr2):
  res = [[0] * N for _ in range(N)]
  for i in range(N):
    for j in range(N):
      for k in range(N):
        res[i][j] += arr1[i][k] * arr2[k][j]
        res[i][j] %= 1000

  return res

# 거듭제곱 계산 함수
def mul(a, b):
  if b == 1:
    return [[x % 1000 for x in row] for row in a]

  half = mul(a, b // 2)
  half_squared = cal(half, half)

  if b % 2 == 1:
    return cal(half_squared, a)
    # cal(mul(a, b // 2), mul(a, b // 2 + 1)) 보다 더 효율적임.
  else:
    return half_squared

N, B = map(int, input().split())
A = [list(map(int, input().split())) for _ in range(N)]

for i in mul(A, B):
  print(*i)
```

```python
# 08 - 메모이제이션 활용
def fibonacci(N, memo = {}):   # memo는 딕셔너리
  if N in memo:
    return memo[N]
  if N == 0:
    return 0
  elif N == 1:
    return 1

  # N이 홀수일 때
  if N % 2 == 1:
    memo[N] = (fibonacci(N // 2 + 1, memo) ** 2 + fibonacci(N // 2, memo) ** 2) % 1000000007
  # N이 짝수일 때
  else:
    memo[N] = (fibonacci(N // 2) * (fibonacci(N // 2) + 2 * fibonacci(N // 2 - 1))) % 1000000007

  return memo[N]

n = int(input())
print(fibonacci(n))
```

```python
# 09 - 분할 정복
def divide_and_conquer(histogram, start, end):
  # 분할 정복을 통해 계속 파고 들어가다보면 1개 또는 2개의 블럭만 남게 되는 때가 온다.
  # 1개의 블록만 남았을 때의 넓이 리턴
  if end == start:
    return histogram[end]
  # 2개의 블록만 남았을 때의 넓이 리턴
  elif end - start == 1:
    if histogram[end] < histogram[start]:
      return max(2 * histogram[end], histogram[start])
    else:
      return max(2 * histogram[start], histogram[end])

  mid = (start + end) // 2
  # 분할 정복을 위해 left, right area로 나누어줌
  left_area = divide_and_conquer(histogram, start, mid)
  right_area = divide_and_conquer(histogram, mid + 1, end)
  left = mid - 1
  right = mid + 1
  mid_area = histogram[mid]
  now_height = histogram[mid]

  # 왼쪽이나 오른쪽 중 하나라도 끝까지 갈 때까지 반복 계산
  while start <= left and right <= end:
    # 오른쪽 블록이 더 높을 때
    if histogram[left] < histogram[right]:
      if histogram[right] < now_height:
        now_height = histogram[right]
      mid_area = max(mid_area, now_height * (right - left))
      right += 1
    # 왼쪽 블록이 더 높거나 같을 때
    else:
      if histogram[left] < now_height:
        now_height = histogram[left]
      mid_area = max(mid_area, now_height * (right - left))
      left -= 1

  # 오른쪽만 끝에 닿았을 때
  while start <= left:
    if histogram[left] < now_height:
      now_height = histogram[left]
    mid_area = max(mid_area, now_height * (right - left))
    left -= 1
  # 왼쪽만 끝에 닿았을 때
  while right <= end:
    if histogram[right] < now_height:
      now_height = histogram[right]
    mid_area = max(mid_area, now_height * (right - left))
    right += 1

  return max(left_area, right_area, mid_area)

res = []
while True:
  histogram = list(map(int, input().split()))

  if histogram[0] == 0:
    break
  n = histogram[0]
  res.append(divide_and_conquer(histogram, 1, n))

for i in res:
  print(i)
```

- 넓이를 구할 때마다 오른쪽에 있는 사각형들을 파고 들어가면서 높이를 따져봐야댐. 오른쪽의 사각형의 높이가 지정한 높이보다 작으면 넓이계산 끝. 이 매커니즘을 재귀함수로 정의해야할듯.

# 아이디어 (완전 탐색, 하지만 이는 시간복잡도가 O(n^2)로 오버됨.)
- 기준 블록을 잡은 뒤에 왼쪽과 오른쪽으로 더 이상 확장할 수 없을 때까지 범위를 넓혀가며 직사각형의 넓이를 계산해준다. 여기서 더 이상 확장할 수 없는 경우란, 각각의 인덱스가 배열의 범위 경곗값에 다다른 경우이거나, 그렇지 않은 경우 인덱스가 가리키는 블록의 높이가 기준 블록의 높이보다 작은 경우이다.

# 아이디어 (분할 정복)
- 중간값을 기준으로 양쪽으로 넓혀가며 최댓값을 찾는 작업을 해보자.
- 예를 들어 (4, 6, 8, 5, 7) 이라고 해보자. 처음에는 중간값인 8에서 시작한다. 이후 왼쪽 또는 오른쪽으로 범위를 넓혀가는데, 둘 중에 높이가 더 높은 쪽으로 확장을 진행한다.
- 첫 번째는 6과 5 중에서 6이 더 크므로 6을 선택, 이 때의 넓이는 6 * 2 = 12 이다. 그 다음으로는 4와 5 중에서 5가 더 크므로 5를 선택, 이 때의 넓이는 5 * 3 = 15이다. 그 다음으로는 4와 7 중에서 7이 더 크므로 넓이는 5 * 4 = 20이다. 마지막으로는 왼쪽의 4로 확장을 진행하는데, 이 경우에는 4 * 5 = 20으로 최대 넓이가 변하지 않기 때문에 넓이의 갱신은 발생하지 않음.

```python
# 09 - 다시 풀어보기

def divide_and_conquer(histogram, start, end):
  if start == end:
    return histogram[start]
  elif end - start == 1:
    if histogram[start] < histogram[end]:
      return max(2 * histogram[start], histogram[end])
    else:
      return max(2 * histogram[end], histogram[start])

  mid = (start + end) // 2
  left_area = divide_and_conquer(histogram, start, mid)
  right_area = divide_and_conquer(histogram, mid + 1, end)
  left = mid - 1
  right = mid + 1
  mid_area = histogram[mid]
  now_height = histogram[mid]

  while start <= left and right <= end:
    if histogram[left] < histogram[right]:
      if histogram[right] < now_height:
        now_height = histogram[right]
      mid_area = max(mid_area, now_height * (right - left))
      right += 1
    else:
      if histogram[left] < now_height:
        now_height = histogram[left]
      mid_area = max(mid_area, now_height * (right - left))
      left -= 1

  while start <= left:
    if histogram[left] < now_height:
      now_height = histogram[left]
    mid_area = max(mid_area, now_height * (right - left))
    left -= 1

  while right <= end:
    if histogram[right] < now_height:
      now_height = histogram[right]
    mid_area = max(mid_area, now_height * (right - left))
    right += 1

  return max(left_area, right_area, mid_area)

results = []
while True:
  histogram = list(map(int, input().split()))

  n = histogram[0]
  if n == 0:
    break
  results.append(divide_and_conquer(histogram, 1, n))

for result in results:
  print(result)
```
