```python
# 02

def fibonacci(num):
  if num <= 1:
    return num
  return fibonacci(num - 1) + fibonacci(num - 2)    # 값을 계산하고 그 결과를 반환해야 하므로 return이 꼭 있어야함.

n = int(input())

print(fibonacci(n))
```

```python
# 04

def merge_sort(array):
  if len(array) <= 1:
    return array

  # 리스트 분할
  mid = (len(array) + 1) // 2
  left_half = merge_sort(array[:mid])
  right_half = merge_sort(array[mid:])    # 쪼갠 각 정렬이 새로운 리스트에 대입 : 추가적인 메모리 공간

  # 병합 과정 : 정렬이 된 하위 배열을 대상으로
  # 더 작은 요소부터 왼쪽에 추가
  i = j = 0
  sorted_array = []
  while i < len(left_half) and j < len(right_half):
    if left_half[i] < right_half[j]:
      sorted_array.append(left_half[i])
      ans.append(left_half[i])
      i += 1
    else:
      sorted_array.append(right_half[j])
      ans.append(right_half[j])
      j += 1

  # 남은 요소 처리
  while i < len(left_half):
    sorted_array.append(left_half[i])
    ans.append(left_half[i])
    i += 1
  while j < len(right_half):
    sorted_array.append(right_half[j])
    ans.append(right_half[j])
    j += 1

  return sorted_array

ans = []
N, K = map(int, input().split())
merge_sort(list(map(int, input().split())))
print(ans[K - 1] if K <= len(ans) else -1)
```

```python
# 06

def rect_cut(start, n):
  if n < 1:
    return
  for a in range(0, N, n * 3):
    for b in range(0, N, n * 3):       # 가로줄
      for i in range(start + n + a, start + n * 2 + a):     # 세로줄
        for j in range(start + n + b, start + n * 2 + b):
          rect_list[i][j] = " "

  if n // 3 > 0:    # if n < 1 조건문 있어서 사실상 필요하진 않지만, 불필요한 호출을 줄여주기 위해 쓴 것.
    rect_cut(start, n // 3)

N = int(input())
#rect_list = [["*"] * N] * N    / 이처럼 나타내면 모든 하위 리스트는 동일한 객체를 참조하게 된다. 따라서 아래처럼 작성해야함.
rect_list = [["*"] * N for _ in range(N)]
rect_cut(0, N // 3)
for i in range(N):
  print("".join(rect_list[i]))
```
