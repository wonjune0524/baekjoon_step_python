```python
# 01

N, M = map(int, input().split())
card_list = list(map(int, input().split()))
card_sum_list = []

for i in range(N - 2):
  for j in range(i + 1, N - 1):
    for k in range(j + 1, N):
      card_sum = card_list[i] + card_list[j] + card_list[k]
      if card_sum <= M:
        card_sum_list.append(card_sum)

print(max(card_sum_list))
```

```python
# 02

N = int(input())
num_list = []

for i in range(10):
  for j in range(10):
    for k in range(10):
      if N == 101 * i + 11 * j + 2 * k:
        num_list.append(100 * i + 10 * j + k)

print(min(num_list))
```

```python
# 02 - 런타임 에러로 인한 다른 풀이

N = int(input())
result = 0

for i in range(1, N + 1):
  num_list = list(map(int, str(i)))
  result = sum(num_list) + i
  if result == N:
    print(i)
    break
  if i == N:
    print(0)
```

```python
# 03 - 분모가 0이 될 때를 고려하지 않아 틀린듯

a, b, c, d, e, f = map(int, input().split())

if a/b == d/e and c/b != f/e:
  print(0)
else:
  print(int((c/b - f/e) / (a/b - d/e)), int((c/a - f/d) / (b/a - e/d)))
```

```python
# 03 - 다른 풀이

a, b, c, d, e, f = map(int, input().split())

x = (c * e - b * f) // (a * e - b * d)
y = (c * d - a * f) // (b * d - a * e)
print(x, y)
```

```python
# 04

N, M = map(int, input().split())
board_list = []
cut_list = []

for i in range(N):
  board_list.append(input())

for p in range(N - 7):
  for i in range(M - 7):
    B_change = 0    # B로 시작했을 때
    W_change = 0    # W로 시작했을 때
    for j in range(p, p + 8):
      for k in range(i, i + 8):
        if (k + j) % 2 == 0:
          if board_list[j][k] != "W":
            W_change += 1
          if board_list[j][k] != "B":
            B_change += 1
        else:
          if board_list[j][k] != "B":
            W_change += 1
          if board_list[j][k] != "W":
            B_change += 1
    cut_list.append(W_change)
    cut_list.append(B_change)

print(min(cut_list))
```

```python
# 06

N = int(input())
count_list = []
count = 0

fir_N = N

while True:
  three_count = 0
  num = fir_N
  num -= 5 * count

  if num < 0:
    break

  if num % 3 == 0:
    three_count += num // 3
    count_list.append(count + three_count)

  count += 1

if len(count_list) == 0:
  print(-1)
else:
  print(min(count_list))
```
