```python
# 01

T = int(input())

def gcd(a, b):
  if a > b:
    a, b = b, a

  for i in range(a, 0, -1):
    if a % i == 0 and b % i == 0:
      break

  return i

for _ in range(T):
  A, B = map(int, input().split())
  print(A * B // gcd(A, B))   # 두 정수를 곱한 값을 최대공약수로 나눠주면 최소공배수.
```

```python
# 01 - 다른 방법으로 풀어보기

import math

T = int(input())

for _ in range(T):
  A, B = map(int, input().split())
  gcd = math.gcd(A, B)

  print(A * B // gcd)
```

```python
# 03

import math

A, B = map(int, input().split())
X, Y = map(int, input().split())

gcd = math.gcd(B, Y)  # 최대공약수
num = B * Y // gcd    # 분수의 분모, 최소공배수

numerator = A * (num // B) + X * (num // Y)   # 분수의 분자

gcd_result = math.gcd(num, numerator)   # 기약분수로 만들기 위해 분자와 분모의 최대공약수 구하기

print(numerator // gcd_result, num // gcd_result)
```

```python
# 04

import math

interval_list = []

N = int(input())
count = 0

dis_list = sorted([int(input()) for _ in range(N)])

for i in range(len(dis_list) - 1):
  interval_list.append(dis_list[i + 1] - dis_list[i])

gcd = interval_list[0]
for i in range(1, len(interval_list)):
  gcd = math.gcd(gcd, interval_list[i])

for i in interval_list:
  count += i // gcd - 1

print(count)
```

```python
# 04 - 다시 풀어보기

import math   # 최대공약수 구하기

N = int(input())
interval_list = []
count = 0

distance_list = sorted([int(input()) for _ in range(N)])

for i in range(len(distance_list) - 1):
  interval_list.append(distance_list[i + 1] - distance_list[i])

gcd = interval_list[0]
for i in range(1, len(interval_list)):
  gcd = math.gcd(gcd, interval_list[i])

for i in interval_list:
  count += i // gcd - 1

print(count)
```

```python
# 05

import math

T = int(input())

for _ in range(T):
  n = int(input())

  if n == 0 or n == 1:
    print(2)
    continue

  while True:
    is_prime = True

    for i in range(2, int(math.sqrt(n)) + 1):    # n의 제곱근까지만 확인하여 코드 효율성 높이기
      if n % i == 0:
        is_prime = False
        break

    if is_prime:
      print(n)
      break

    n += 1
```

```python
# 05 - n보다 같다는 조건을 못봤음

import math

def check(num):
  for i in range(2, int(math.sqrt(num)) + 1):
    if num % i == 0:
      return False
  return True

T = int(input())

for i in range(T):
  n = int(input())

  """
  if n == 0 or n == 1:
    print(2)
    continue
  """

  while True:
    if n == 0 or n == 1:
      print(2)
      break
    if check(n):
      print(n)
      break
    else:
      n += 1
```

```python
# 06

import math

M, N = map(int, input().split())

for i in range(M, N + 1):
  is_prime = True

  if i == 1:
    continue

  for j in range(2, int(math.sqrt(i)) + 1):
    if i % j == 0:
      is_prime = False
      break

  if is_prime:
    print(i)
```

```python
# 07

import math

while True:
  n = int(input())
  count = 0

  if n == 0:
    break

  for i in range(n + 1, 2 * n + 1):
    is_prime = True

    """
    n이 자연수여서 이 조건이 실행될 일이 없긴함
    if i == 1:
      continue
    """

    for j in range(2, int(math.sqrt(i)) + 1):
      if i % j == 0:
        is_prime = False
        break

    if is_prime:
      count += 1

  print(count)
```

```python
# 07 - 시간 초과로 인한 다른 풀이(에라토스테네스의 체, 계수 정렬이랑 유사함)

check = [0] * 2 + [1] * 246911

# 첫 소수만 1로 남기고 소수의 배수들은 소수가 아니므로 0으로 초기화해줌.
for i in range(2, 246913):
  if check[i]:
    for j in range(i * 2, 246913, i):
      check[j] = 0

while True:
  n = int(input())

  if n == 0:
    break

  print(sum(check[n + 1:2*n + 1]))
```

```python
# 08 - 에라토스테네스의 체 사용

check = [0] * 2 + [1] * 999999

for i in range(2, 1000001):
  if check[i]:
    for j in range(i * 2, 1000001, i):
      check[j] = 0

T = int(input())

for i in range(T):
  N = int(input())
  count = 0

  for j in range(2, N // 2 + 1):
    if check[j]:
      if check[N - j]:
        count += 1

  print(count)
```

```python
# 09

N = int(input())

check = [0] * (N + 1)

for i in range(1, N + 1):
  for j in range(i, N + 1, i):
    if check[j] == 0:
      check[j] = 1
    elif check[j] == 1:
      check[j] = 0

print(sum(check))
```
