```python
# 03

x_num = []
y_num = []
result_x = 0
result_y = 0

for i in range(3):
  x , y = map(int, input().split())
  x_num.append(x)
  y_num.append(y)

for i in range(3):
  if x_num.count(x_num[i]) == 1:
    result_x = x_num[i]
  if y_num.count(y_num[i]) == 1:
    result_y = y_num[i]

print(result_x, result_y)
```

```python
# 05

x_list = []
y_list = []
N = int(input())

for i in range(N):
  x, y = map(int, input().split())
  x_list.append(x)
  y_list.append(y)

print((max(x_list) - min(x_list)) * (max(y_list) - min(y_list)))
```

```python
# 06

fst_angle = int(input())
sec_angle = int(input())
trd_angle = int(input())
sum = fst_angle + sec_angle + trd_angle

if fst_angle == sec_angle == trd_angle == 60:
  print("Equilateral")
elif sum == 180:
  if fst_angle != sec_angle and sec_angle != trd_angle and trd_angle != fst_angle:
    print("Scalene")
  else:
    print("Isosceles")
else:
  print("Error")
```

```python
# 07

while True:
  length_list = []
  a, b, c = map(int, input().split())
  length_list.append(a)
  length_list.append(b)
  length_list.append(c)

  if a == b == c == 0:
    break
  elif max(length_list) < sum(length_list) - max(length_list):
    if a == b == c:
      print("Equilateral")
    elif a != b and b != c and c != a:
      print("Scalene")
    else:
      print("Isosceles")
  else:
    print("Invalid")
```

```python
# 08

length_list = []
a, b, c = map(int, input().split())
length_list.append(a)
length_list.append(b)
length_list.append(c)
rest_sum = sum(length_list) - max(length_list)

if max(length_list) < rest_sum:
  print(sum(length_list))
else:
  length_list[length_list.index(max(length_list))] = rest_sum - 1
  print(sum(length_list))
```
