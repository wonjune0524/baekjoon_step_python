```python
# 02

N = int(input())
name_set = set()
emo_num = 0

for i in range(N):
  data = input()

  if "ENTER" in name_set:
    if data == "ENTER":
      emo_num += len(name_set) - 1
      name_set.clear()

  name_set.add(data)

if emo_num == 0:
  print(len(name_set) - 1)
else:
  print(len(name_set) - 1 + emo_num)
```

```python
# 03

N = int(input())
dic = {}

for i in range(N):
  A, B = input().strip().split()

  if A not in dic:
    dic[A] = 0
  if B not in dic:
    dic[B] = 0

  if A == "ChongChong" or dic[A] == 1:
    dic[A] = 1
    dic[B] = 1
  elif B == "ChongChong" or dic[B] == 1:
    dic[A] = 1
    dic[B] = 1

print(sum(dic.values()))
```

```python
# 05

N, M = map(int, input().split())
data_dic = dict()

for i in range(N):
  data = input()

  if len(data) >= M:
    if data not in data_dic:
      data_dic[data] = 0
    data_dic[data] += 1

sorted_items = sorted(data_dic.items(), key = lambda data : (-data[1], -len(data[0]), data[0]))

for i in range(len(sorted_items)):
  print(sorted_items[i][0])
```
