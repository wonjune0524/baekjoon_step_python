```python
# 01

N = int(input())
stack_list = []

for i in range(N):
  data = input().split()

  if data[0] == "1":
    stack_list.append(int(data[-1]))
  elif data[0] == "2":
    if stack_list:
      print(stack_list.pop(-1))
      continue
    print(-1)
  elif data[0] == "3":
    print(len(stack_list))
  elif data[0] == "4":
    if stack_list:
      print(0)
      continue
    print(1)
  elif data[0] == "5":
    if stack_list:
      print(stack_list[-1])
      continue
    print(-1)
```

```python
# 01 - 다시 풀어보기

N = int(input())
stack_list = []

for _ in range(N):
  data = input().split()

  if data[0] == "1":
    stack_list.append(int(data[-1]))
  elif data[0] == "2":
    if stack_list:
      print(stack_list.pop())
      continue
    print(-1)
  elif data[0] == "3":
    print(len(stack_list))
  elif data[0] == "4":
    if stack_list:
      print(0)
      continue
    print(1)
  elif data[0] == "5":
    if stack_list:
      print(stack_list[-1])
      continue
    print(-1)
```

```python
# 02

K = int(input())
stack_list = []

for i in range(K):
  data = int(input())

  if data == 0:
    stack_list.pop()
    continue
  stack_list.append(data)

print(sum(stack_list))
```

```python
# 03

T = int(input())

for _ in range(T):
  stack_list = []
  is_invalid = False
  text = input().strip()

  for i in text:
    if i == "(":
      stack_list.append(i)
    elif i == ")":
      if stack_list:
        stack_list.pop()
        continue
      is_invalid = True
      break

  if stack_list or is_invalid:
    print("NO")
    continue
  print("YES")
```

```python
# 04

while True:
  text = input()
  stack_list = []
  is_invalid = True

  if text == ".":
    break

  for i in text:
    if i in "([":
      stack_list.append(i)
    elif i == ")":
      if stack_list and stack_list[-1] == "(":
        stack_list.pop()
        continue
      is_invalid = False
      break
    elif i == "]":
      if stack_list and stack_list[-1] == "[":
        stack_list.pop()
        continue
      is_invalid = False
      break

  if stack_list or not is_invalid:
      print("no")
      continue
  print("yes")
```

```python
# 05 - 2 1 3 / 5 4 2 3 1 과 같은 경우를 고려하자

N = int(input())
stack_list = []
snack_list = [0]

num_list = list(map(int, input().split()))

while num_list:
  is_invalid = True
  if not stack_list and num_list[0] != (snack_list[-1] + 1):
    stack_list.append(num_list.pop(0))

  elif num_list[0] == (snack_list[-1] + 1):
    snack_list.append(num_list.pop(0))

  else:
    stack_list.append(num_list.pop(0))    # 위 조건 외의 숫자들은 마구잡이로 다 집어넣는다.

  while is_invalid:
    if stack_list and stack_list[-1] == (snack_list[-1] + 1):
      snack_list.append(stack_list.pop())
      continue
    is_invalid = False

if stack_list:
  print("Sad")
else:
  print("Nice")
```

```python
# 05 - 다른 풀이 / 리스트를 하나만 써보기

N = int(input())
num_list = list(map(int, input().split()))
stack_list = []
snack_num = 1

for num in num_list:
  is_invalid = True

  if num == snack_num:
    snack_num += 1
  else:
    stack_list.append(num)

  while is_invalid:
    if stack_list and stack_list[-1] == snack_num:
      stack_list.pop()
      snack_num += 1
      continue
    is_invalid = False

if stack_list:
  print("Sad")
else:
  print("Nice")
```

```python
# 06

from collections import deque

N = int(input())
queue_list = deque()

for _ in range(N):
  data = input().split()

  if data[0] == "push":
    queue_list.append(int(data[-1]))
  elif data[0] == "pop":
    if queue_list:
      print(queue_list.popleft())
      continue
    print(-1)
  elif data[0] == "size":
    print(len(queue_list))
  elif data[0] == "empty":
    if queue_list:
      print(0)
      continue
    print(1)
  elif data[0] == "front":
    if queue_list:
      print(queue_list[0])
      continue
    print(-1)
  elif data[0] == "back":
    if queue_list:
      print(queue_list[-1])
      continue
    print(-1)
```

```python
# 07

from collections import deque

N = int(input())
queue_list = deque()

for i in range(1, N + 1):
  queue_list.append(i)

while len(queue_list) > 1:
  queue_list.popleft()    # deque에서 pop은 항상 맨오른쪽의 요소만 제거하므로 pop(0)과 같은 사용은 불가능함.
  queue_list.append(queue_list.popleft())

print(queue_list[0])
```

```python
# 09

from collections import deque

N = int(input())
deque_list = deque()

for _ in range(N):
  data = input().split()

  if data[0] == "1":
    deque_list.appendleft(int(data[1]))
  elif data[0] == "2":
    deque_list.append(int(data[1]))
  elif data[0] == "3":
    if len(deque_list) == 0:
      print(-1)
    else:
      print(deque_list.popleft())
  elif data[0] == "4":
    if len(deque_list) == 0:
      print(-1)
    else:
      print(deque_list.pop())
  elif data[0] == "5":
    print(len(deque_list))
  elif data[0] == "6":
    if len(deque_list) == 0:
      print(1)
    else:
      print(0)
  elif data[0] == "7":
    if len(deque_list) == 0:
      print(-1)
    else:
      print(deque_list[0])
  elif data[0] == "8":
    if len(deque_list) == 0:
      print(-1)
    else:
      print(deque_list[-1])
```

```python
# 10

from collections import deque

N = int(input())
deque_list = deque(enumerate(map(int, input().split())))
answer_list = []

while len(deque_list) > 0:
  idx, next_index = deque_list.popleft()
  answer_list.append(idx + 1)

  if next_index > 0:
    deque_list.rotate(-next_index + 1)
  elif next_index < 0:
    deque_list.rotate(-next_index)

print(" ".join(map(str, answer_list)))
```

```python
# 11

# stack에 원소를 넣고 pop을 하게 되면 넣은 원소를 그대로 pop하는 것이므로 의미가 없다. 그냥 건너뛰면 됨.
# queue는 원소를 앞에서 넣고 pop(큐에서 pop은 맨 뒤의 원소를 삭제함)함.
from collections import deque

N = int(input())
A_list = list(map(int, input().split()))
B_list = list(map(int, input().split()))
M = int(input())
C_list = list(map(int, input().split()))

deque_list = deque()

for i in range(N):
  if A_list[i] == 0:
    deque_list.appendleft(B_list[i])

for i in C_list:
  deque_list.append(i)
  print(deque_list.popleft(), end = " ")
```
