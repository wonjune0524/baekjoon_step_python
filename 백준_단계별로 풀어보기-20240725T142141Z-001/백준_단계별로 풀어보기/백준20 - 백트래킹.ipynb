```python
# 05

def sol(k): # k : 놓은 말 개수 - 점유 행 번호로도 활용 가능
    # 퀸 : 상하좌우 - x, y 좌표 확인 / 대각선 - (y=x 선상 : used_up) r+c 같으면. (y=-x 선상 : used_down) r-c같으면 놓을 수 없음.
    global n, cnt

    if k == n :
        cnt += 1
        return

    # i : 점유 열 번호로 활용 가능
    for i in range(n):
        if not used_c[i] and not used_up[k+i] and not used_down[(n-1)+k-i]:
            used_c[i] = True
            used_up[k+i] = True
            used_down[(n-1)+k-i] = True
            sol(k+1)
            used_c[i] = False
            used_up[k+i] = False
            used_down[(n-1)+k-i] = False

## input
n = int(input())
maps = [[0]*n for _ in range(n)]
# 열 점유 여부
used_c = [False]*n
# y = x 선상 점유 여부
used_up = [False]*(2*(n-1)+1)
# y = -x 선상 점유 여부
used_down = [False]*(2*(n-1)+1)

cnt = 0

## output
sol(0)
print(cnt)
```

```python
# 07 - 다른 풀이

N = int(input())

num_list = list(map(int, input().split()))
oper_list = list(map(int, input().split()))

num_max = -1e9    # 이는 -(1 * 10^9 를 의미함)
num_min = 1e9

def dfs(num, temp):
  global num_max, num_min

  if num == N - 1:
    num_max = max(temp, num_max)
    num_min = min(temp, num_min)
    return

  if oper_list[0] != 0:   # 덧셈
    oper_list[0] -= 1
    dfs(num + 1, temp + num_list[num + 1])
    oper_list[0] += 1

  if oper_list[1] != 0:   # 뺄셈
    oper_list[1] -= 1
    dfs(num + 1, temp - num_list[num + 1])
    oper_list[1] += 1

  if oper_list[2] != 0:    # 곱셈
    oper_list[2] -= 1
    dfs(num + 1, temp * num_list[num + 1])
    oper_list[2] += 1

  if oper_list[3] != 0:   # 나눗셈
    oper_list[3] -= 1
    dfs(num + 1, int(temp / num_list[num + 1]))
    oper_list[3] += 1

dfs(0, num_list[0])
print(num_max)
print(num_min)
```

```python
# 07 - 다시 풀어보기

N = int(input())
num_list = list(map(int, input().split()))
oper_list = list(map(int, input().split()))

max_num = -1e9
min_num = 1e9

def dfs(num, temp):
  global max_num, min_num

  if num == N - 1:
    max_num = max(max_num, temp)
    min_num = min(min_num, temp)
    return

  if oper_list[0] != 0:
    oper_list[0] -= 1
    dfs(num + 1, temp + num_list[num + 1])
    oper_list[0] += 1

  if oper_list[1] != 0:
    oper_list[1] -= 1
    dfs(num + 1, temp - num_list[num + 1])
    oper_list[1] += 1

  if oper_list[2] != 0:
    oper_list[2] -= 1
    dfs(num + 1, temp * num_list[num + 1])
    oper_list[2] += 1

  if oper_list[3] != 0:
    oper_list[3] -= 1
    dfs(num + 1, int(temp / num_list[num + 1]))
    oper_list[3] += 1

dfs(0, num_list[0])
print(max_num)
print(min_num)
```

```python
# 08 - 내 풀이

N = int(input())
team_list = []        # 선수들 S 리스트
count_arr = [1] + [0] * (N - 1)   # 0, 1 값을 기준으로 두 팀으로 나눠주는 리스트
sep_arr1 = []                # 선수 번호 분리 리스트
sep_arr2 = []
result_list = []

# 팀 능력치 입력받기
for _ in range(N):
  team_list.append(list(map(int, input().split())))

# 두 선수의 능력치 합 계산 함수
def cal(sep1, sep2):
  return team_list[sep1][sep2] + team_list[sep2][sep1]

# 팀의 능력치 계산 함수
def score(num):
  result1 = 0
  result2 = 0
  if len(sep_arr1) == 2:
    result1 += cal(sep_arr1[0], sep_arr1[1])
    return result1
  if len(sep_arr2) == 2:
    result2 += cal(sep_arr2[0], sep_arr2[1])
    return result2

  for i in range(num, N):
    if count_arr[i] == 0:
      sep_arr1.append(i)
      result1 += score(num + 1)
      sep_arr1.pop()
    if count_arr[i] == 1:
      sep_arr2.append(i)
      result2 += score(num + 1)
  return abs(result1- result2)

def dfs_team(start):
  if start == N // 2:
    result_list.append(score(0))
    return

  for i in range(start, N):
    count_arr[i] += 1
    dfs_team(start + 1)
    count_arr[i] -= 1

dfs_team(1)
print(result_list)
```

```python
# 08 - gpt 도움

N = int(input())
team_list = []        # 선수들 S 리스트
result_list = []

# 팀 능력치 입력받기
for _ in range(N):
  team_list.append(list(map(int, input().split())))

# 팀 능력치 계산 함수
def cal(team):
  total = 0
  for i in range(len(team)):
    for j in range(i + 1, len(team)):
      total += team_list[team[i]][team[j]] + team_list[team[j]][team[i]]
  return total

def dfs_team(index, team1, team2):
  global result_list    # 없어도 됨

  # 모든 선수를 다 분배했을 때
  if index == N:
    if len(team1) == N // 2 and len(team2) == N // 2:
      score1 = cal(team1)
      score2 = cal(team2)
      result_list.append(abs(score1 - score2))
    return

  # 팀 1
  if len(team1) < N // 2:
    dfs_team(index + 1, team1 + [index], team2)

  # 팀 2
  if len(team2) < N // 2:
    dfs_team(index + 1, team1, team2 + [index])

dfs_team(0, [], [])

print(min(result_list))
```
