```python
# 01 - 시간 초과로 인한 다른 풀이

N = int(input())
num_list = list(map(int, input().split()))

M = int(input())
correct_list = list(map(int, input().split()))

dic = {}
for i in correct_list:
  dic[i] = 0

for i in num_list:
  if i in dic:
    dic[i] = 1

for i in dic:
  print(dic[i], end = " ")
```

```python
# 02 - 집합(set)으로 풀기

N, M = map(int, input().split())
S_set = set()
count = 0

for _ in range(N):
  data = input().rstrip()   # rstrip 메서드는 불필요한 우측 공백이나 줄바꿈 문자를 제거해야 할 때 유용.
  S_set.add(data)           # set은 add로 데이터를 추가함.

for _ in range(M):
  data = input().rstrip()
  if data in S_set:
    count += 1

print(count)
```

```python
# 02 - 딕셔너리(dic)로 풀기

dic = {}
count = 0
N, M = map(int, input().split())

for _ in range(N):
  data = input().rstrip()
  dic[data] = True

for _ in range(M):
  data = input().rstrip()
  if data in dic:
    count += 1

print(count)
```

```python
# 03

n = int(input())
dic = {}
name_list = []

for i in range(n):
  name, status = input().rstrip().split()
  dic[name] = status

for i in dic:
  if dic[i] == "enter":
    name_list.append(i)

name_list.sort(reverse = True)

for i in name_list:
  print(i)
```

```python
# 03 - 좀 더 간단한 방법

n = int(input())
dic = {}

for i in range(n):
  name, status = input().rstrip().split()
  dic[name] = status
  if dic[name] == "leave":
    del dic[name]

"""
이런 식으로 쓰면 반복 중에 dic의 사이즈가 줄어들기 때문에 안됨.
for i in dic:
  if dic[i] == "leave":
    del dic[i]
"""

d = sorted(dic.items(), reverse = True)   # items 메서드는 딕셔너리의 각 항목을 (키, 값) 형태의 튜플로 반환해준다.
dic = dict(d)

for i in dic.keys():
  print(i)
```

```python
# 04

N, M = map(int, input().split())
num_dic = {}
poc_dic = {}    # key에 접근은 반복문으로 해야하기 때문에 2개의 딕셔너리로 나눠준다.

for i in range(1, N + 1):
  data1 = input()
  num_dic[data1] = i
  poc_dic[i] = data1

for _ in range(M):
  data2 = input()

  if data2.isalpha():
    print(num_dic[data2])
  else:
    print(poc_dic[int(data2)])
```

```python
# 05

N = int(input())
num_list = list(map(int, input().split()))

M = int(input())
correct_list = list(map(int, input().split()))

dic = {i: 0 for i in correct_list}

for i in num_list:
  if i in dic:
    dic[i] += 1

for i in correct_list:
  print(dic[i], end = " ")
```

```python
# 06

N, M = map(int, input().split())
dic = {}
count = 0
print_list = []

for _ in range(N):
  data1 = input().rstrip()
  dic[data1] = True

for _ in range(M):
  data2= input().rstrip()
  if data2 in dic:
    count += 1
    print_list.append(data2)

print_list.sort()

print(count)    # count 쓰지말고 그냥 len(print_list) 하면 될듯
for i in print_list:
  print(i)
```

```python
# 06 - 집합(set)으로 풀어보기

# import sys

# input = sys.stdin.readline

N, M = map(int, input().split())

set1 = set()
set2 = set()

# 듣도 못한 사람
for _ in range(N):
  set1.add(input().rstrip())

# 보도 못한 사람
for _ in range(M):
  set2.add(input().rstrip())

# 듣보잡들
result = sorted(list(set1 & set2))

# 듣보잡 수
print(len(result))

# 듣보잡 이름
for i in result:
  print(i)
```

```python
# 08

data = input()
data_set = set()

for i in range(1, len(data) + 1):
  for j in range(len(data) - (i - 1)):
    data_set.add(data[j:j + i])

print(len(data_set))
```
